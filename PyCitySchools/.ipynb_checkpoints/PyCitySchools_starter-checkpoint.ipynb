```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#print(f'{school_data.head(20)}')
#print(f'{student_data.head()}')
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
tot_stu = student_data["Student ID"].count()
avg_math = student_data["math_score"].mean()
avg_read = student_data["reading_score"].mean()
ovr_avg_score = (avg_math + avg_read) / 2

pass_math_stu = student_data.loc[student_data["math_score"] >= 70, :]
cnt_math_pass = pass_math_stu["Student ID"].count()
#print(f'{cnt_math_pass}')
pass_read_stu = student_data.loc[student_data["reading_score"] >= 70, :]
cnt_read_pass = pass_read_stu["Student ID"].count()
#print(f'{cnt_read_pass}')
Dist_sum = pd.DataFrame({"Total Schools" : [school_data["school_name"].count()], 
                         "Total Students" : [tot_stu],
                         "Total Budget" : [school_data["budget"].sum()],
                         "Average Math Score" : [avg_math],
                         "Average Reading Score" : [avg_read],
                         "% Passing Math" : [(cnt_math_pass / tot_stu) * 100],
                         "% Passing Reading" : [(cnt_read_pass / tot_stu) * 100],
                         "% Overall Passing Rate" : [ovr_avg_score],
                        })
Dist_sum["Total Students"] = Dist_sum["Total Students"].map("{:,}".format)
Dist_sum["Total Budget"] = Dist_sum["Total Budget"].map("${:,}".format)
Dist_sum
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Creatin the school summary

# Converting the merged data to a Dataframe
df = pd.DataFrame(school_data_complete)

# Creating a dataframe with students having pass marks in math
df_math = df[["school_name", "math_score"]]
df_math_pass = df_math.loc[df_math["math_score"] >= 70, :]
math_df = pd.DataFrame(df_math_pass.groupby("school_name").agg({"math_score":'count'}))

math_df_new = pd.DataFrame({"school_name" : math_df.index, 
                           "math_cnt" : math_df["math_score"]
                           })
#print(math_df_new.head())
# Creating a dataframe with students having pass marks in reading
df_read = df[["school_name", "reading_score"]]
df_read_pass = df_read.loc[df_read["reading_score"] >= 70, :]
read_df = pd.DataFrame(df_read_pass.groupby("school_name").agg({"reading_score":'count'}))
read_df_new = pd.DataFrame({"school_name" : read_df.index, 
                           "read_cnt" : read_df["reading_score"]
                           })
#print(read_df_new.head())

# Combine the data the math and reading counts into a single dataframe
math_read_df = pd.merge(math_df_new, read_df_new, on=["school_name", "school_name"])
#math_read_df
school_type = df[["school_name", "type"]]
# creating the group by on the school name from the merged student and school dataframe
df1 = pd.DataFrame (df.groupby('school_name').aggregate({"Student ID":'count',
                                                        "budget":"mean",
                                                        "math_score":"mean",
                                                        "reading_score":"mean"
                                                       })
                    )
df1["Avg Student Budget"] = df1["budget"] / df1["Student ID"]
df1["school_name"] = df1.index
#print(df1.head())

# Combine the group by dataframe and the math and reading count dataframe. The % pass will be calculated from this
comb_df1 = pd.merge(math_read_df, df1, on=["school_name", "school_name"])
comb_df = pd.merge(comb_df1, school_type, on=["school_name", "school_name"])
comb_df = comb_df.rename(columns={"Student ID":"Total Students", "Avg Student Budget":"Per Student Budget", "budget":"Total School Budget", "math_score":"Avg math score","reading_score":"Avg reading score"})
comb_df["% Passing Math"] = comb_df["math_cnt"] / comb_df["Total Students"] * 100
comb_df["% Passing Reading"] = comb_df["read_cnt"] / comb_df["Total Students"] * 100
comb_df["Overall Passing Rate"] = (comb_df["% Passing Math"] + comb_df["% Passing Reading"]) / 2
#print(comb_df.head())

#Including only the required columns for the display
sch_grp_df = pd.DataFrame ({"school_name":comb_df["school_name"],
                            "type" : comb_df["type"],
                           "Total Students":comb_df["Total Students"], 
                           "Total School Budget":comb_df["Total School Budget"], 
                           "Per Student Budget":comb_df["Per Student Budget"], 
                           "Avg math score":comb_df["Avg math score"], 
                           "Avg reading score":comb_df["Avg reading score"], 
                           "% Passing Math":comb_df["% Passing Math"], 
                           "% Passing Reading": comb_df["% Passing Reading"], 
                           "Overall Passing Rate":comb_df["Overall Passing Rate"]
                           })
sch_grp_df["Total School Budget"] = sch_grp_df["Total School Budget"].map("${:,}".format)
sch_grp_df["Per Student Budget"] = sch_grp_df["Per Student Budget"].map("${:,}".format)
sch_grp_df.head()
```

```python
# Top Schools
top_sch = sch_grp_df.sort_values("Overall Passing Rate", ascending=False)
top_sch.head()
```

## Math Scores by Grade

```python
grade_df = df[["school_name", "math_score","grade"]]
df09 = grade_df.loc[(grade_df["grade"] == '9th')] 
df09_mean = df09.groupby("school_name")["math_score"].sum() / df09.groupby("school_name")["grade"].count()

df10 = grade_df.loc[(grade_df["grade"] == '10th')] 
df10_mean = df10.groupby("school_name")["math_score"].sum() / df10.groupby("school_name")["grade"].count()

df11 = grade_df.loc[(grade_df["grade"] == '11th')] 
df11_mean = df11.groupby("school_name")["math_score"].sum() / df11.groupby("school_name")["grade"].count()

df12 = grade_df.loc[(grade_df["grade"] == '12th')] 
df12_mean = df12.groupby("school_name")["math_score"].sum() / df12.groupby("school_name")["grade"].count()

math_mean_df = pd.DataFrame({"School name":grade_df.groupby("school_name")["school_name"].first(),
                             "9th":df09_mean,
                             "10th":df10_mean,
                             "11th":df11_mean,
                             "12th":df12_mean
                            })
math_mean_df.set_index(keys="School name",inplace=True)
math_mean_df.head()
```
